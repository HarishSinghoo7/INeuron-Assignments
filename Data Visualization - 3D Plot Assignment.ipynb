**Importing required libraries**

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from sklearn import decomposition
from sklearn import datasets
%matplotlib inline
```

**Getting iris dataset**

```python
iris = datasets.load_iris()
```

**Dimensionality Reduction**

```python
pca_dr = decomposition.PCA(n_components=3)
X = pca_dr.fit_transform(iris.data)
X.shape
```

```python
y = iris.target
```

**Plotting 3D graph for new dimensions**

```python
plt.figure(figsize=(12,8))
ax=plt.subplot(111, projection='3d')
sc = Axes3D.scatter3D(ax,xs=X[:,0],ys=X[:,1],zs=X[:,2], c=y, cmap="Spectral", edgecolor='k', s=40)

ax.set_xlabel('X-Axis')
ax.set_ylabel('Y-Axis')
ax.set_zlabel('Z-Axis')

labelTups = [('Setosa', 0), ('Versicolour', 1), ('Virginica', 2)]

colors = [sc.cmap(sc.norm(i)) for i in [1, 2, 0]]

custom_lines = [plt.Line2D([],[], ls="", marker='.', 
                mec='k', mfc=c, mew=.1, ms=20) for c in colors]

ax.legend(custom_lines, [lt[0] for lt in labelTups], 
          loc='center left', bbox_to_anchor=(1.0, .5))

plt.show()
```
